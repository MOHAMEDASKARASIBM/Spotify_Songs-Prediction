```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
#it is use to check the null value of each columns
pd.isnull(df_tracks).sum()
```

```python
#it is used to see how many values are present in the dataset
#what is data type of the columns then how much space it takes
df_tracks.info()
```

```python
#it is gives the popular songs in spotify
sorted_df=df_tracks.sort_values("popularity",ascending=False).head(10)
sorted_df
```

```python
#descriptive analysis
df_tracks.describe().transpose()
#it is used to find important values such as mean,standard division,min and max then percentaile values
```

```python
#set_index function use to set my index as realease time of the song.
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#it use to locate the artist in a specific position
#it is get the name of first ten artists
df_tracks[["artists"]].iloc[:10]
```

```python
#Convert the milesecond duration into mintues
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
#it changes the duration to mintues
df_tracks.duration
```

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correaltion Heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#Get only 4% of data as a sample
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
#we can plot the regression line of loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="blue").set(title="Loudness vs Energy Correlation")
#it highly correalated if energy is increses the loudness also increase
```

```python
#it represent the correlation between the popularity vs dance ability
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="danceability",color="c").set(title="Popularity vs Dance ability Correlation")
```

```python
df_tracks["dates"]=df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
# pip install --user seaborn==0.11.0
```

```python
sns.displot(years,discrete=True,aspect=2,height=7,kind="hist").set(title="Number of songs per year")
```

```python
total_duration=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False).set(title="years vs duration ")
plt.xticks(rotation=90)
#it is used to analyse where the song duration is increase in which period.
```

```python
total_duration=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_duration,ax=ax).set(title="years vs duration ")
plt.xticks(rotation=60)
```
